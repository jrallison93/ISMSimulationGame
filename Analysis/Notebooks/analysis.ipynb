```python
import sys
from pathlib import Path

def find_project_root(markers=("src", "Data", "Notebooks")):
    current_path = Path.cwd().resolve()

    for path in [current_path] + list(current_path.parents):
        has_all_markers = all((path / marker).exists() for marker in markers)
        if has_all_markers:
            return path

    raise RuntimeError(
        f"Project root could not be found from: {current_path}"
    )

project_root = find_project_root()
src_path = project_root / "src"

if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))

import paths as P
```

```python
import pandas as pd

merged_file = P.cleaned_dir / "merged_cleaned_data.xlsx"
merged_data = pd.read_excel(merged_file)

print("Merged data shape:")
print(merged_data.shape)
print()

print("First 10 columns:")
print(merged_data.columns.tolist()[:10])
```

```python
# define the 20 items under the five constructs

construct_items = {
    "cognitive_decision_making_confidence": [
        "i_can_identify_the_most_relevant_security_options_available_when_making_a_decision",
        "i_can_understand_the_potential_benefits_of_different_security_choices",
        "i_can_recognise_the_risks_or_vulnerabilities_associated_with_security_options_available_to_me",
        "i_can_interpret_technical_information_well_enough_to_make_an_informed_decision",
        "i_can_determine_which_security_choice_best_fits_the_needs_of_my_environment_or_organisation"
    ],
    
    "behavioural_decision_making_confidence": [
        "i_can_make_security_related_decisions_without_feeling_uncertain",
        "i_can_express_concerns_about_different_security_options_effectively",
        "i_can_seek_advice_from_security_experts_or_colleagues_when_needed",
        "i_can_handle_pressure_when_making_security_decisions",
        "i_can_communicate_my_security_decisions_effectively_to_stakeholders"
    ],
    
    "cia_triad": [
        "i_can_define_the_confidentiality_component_of_the_cia_triad",
        "i_can_define_the_integrity_component_of_the_cia_triad",
        "i_can_define_the_availability_component_of_the_cia_triad"
    ],
    
    "access_control": [
        "i_can_explain_authentication_in_access_control",
        "i_can_explain_authorization_in_access_control"
    ],
    
    "malware_and_hacking": [
        "i_can_identify_different_types_of_malware_that_can_affect_information_security",
        "i_can_explain_how_malware_can_infiltrate_a_system",
        "i_can_describe_how_hacking_attacks_can_compromise_information_security",
        "i_can_identify_at_least_two_different_types_of_vulnerabilities_in_a_system",
        "i_can_apply_my_knowledge_of_attacks_to_make_informed_decisions_about_information_security"
    ]
}

for construct, items in construct_items.items():
    print(construct, ":", len(items), "items")
```

```python
# check that all expected pre and post item columns exist in merged_data

missing_columns = []

for construct, items in construct_items.items():
    for item in items:
        pre_col = "pre_" + item
        post_col = "post_" + item
        
        if pre_col not in merged_data.columns:
            missing_columns.append(pre_col)
        if post_col not in merged_data.columns:
            missing_columns.append(post_col)

print("Missing columns:")
print(missing_columns)
print()

print("Number of missing columns:")
print(len(missing_columns))
```

```python
# item-level pre and post mean and standard deviation

item_results = []

for construct, items in construct_items.items():
    for item in items:
        pre_col = "pre_" + item
        post_col = "post_" + item
        
        pre_mean = merged_data[pre_col].mean()
        pre_sd = merged_data[pre_col].std()
        post_mean = merged_data[post_col].mean()
        post_sd = merged_data[post_col].std()
        
        item_results.append({
            "construct": construct,
            "item": item,
            "pre_mean": round(pre_mean, 3),
            "pre_sd": round(pre_sd, 3),
            "pre_mean_sd": f"{pre_mean:.2f} ({pre_sd:.2f})",
            "post_mean": round(post_mean, 3),
            "post_sd": round(post_sd, 3),
            "post_mean_sd": f"{post_mean:.2f} ({post_sd:.2f})"
        })

item_results = pd.DataFrame(item_results)

print("Item-level results:")
display(item_results)
```

```python
# save item-level pre/post means and standard deviations
item_results.to_csv(P.tables_dir / "item_pre_post_means_sd.csv", index=False)

print("Saved item_pre_post_means_sd.csv")
print("Shape:", item_results.shape)
```

```python
# create construct-level pre and post mean scores for each student

for construct, items in construct_items.items():
    pre_cols = ["pre_" + item for item in items]
    post_cols = ["post_" + item for item in items]
    
    merged_data["pre_" + construct] = merged_data[pre_cols].mean(axis=1)
    merged_data["post_" + construct] = merged_data[post_cols].mean(axis=1)

construct_score_cols = [
    col for col in merged_data.columns
    if col.startswith("pre_") and col.replace("pre_", "", 1) in construct_items
] + [
    col for col in merged_data.columns
    if col.startswith("post_") and col.replace("post_", "", 1) in construct_items
]

print("Construct score columns created:")
print(construct_score_cols)
```

```python
# construct-level pre and post mean and standard deviation

construct_results = []

for construct in construct_items.keys():
    pre_col = "pre_" + construct
    post_col = "post_" + construct
    
    pre_mean = merged_data[pre_col].mean()
    pre_sd = merged_data[pre_col].std()
    post_mean = merged_data[post_col].mean()
    post_sd = merged_data[post_col].std()
    
    construct_results.append({
        "construct": construct,
        "n_complete_pairs": merged_data[[pre_col, post_col]].dropna().shape[0],
        "pre_mean": round(pre_mean, 3),
        "pre_sd": round(pre_sd, 3),
        "pre_mean_sd": f"{pre_mean:.2f} ({pre_sd:.2f})",
        "post_mean": round(post_mean, 3),
        "post_sd": round(post_sd, 3),
        "post_mean_sd": f"{post_mean:.2f} ({post_sd:.2f})"
    })

construct_results = pd.DataFrame(construct_results)

display(construct_results)
```

```python
# save construct-level pre/post means and standard deviations
construct_results.to_csv(P.tables_dir / "construct_pre_post_means_sd.csv", index=False)

print("Saved construct_pre_post_means_sd.csv")
print("Shape:", construct_results.shape)
```

```python
# Cronbach alpha for each construct at pre and post

def cronbach_alpha(df):
    item_scores = df.dropna().copy()
    k = item_scores.shape[1]

    if k < 2:
        return pd.NA

    item_variances = item_scores.var(axis=0, ddof=1)
    total_score = item_scores.sum(axis=1)
    total_variance = total_score.var(ddof=1)

    if total_variance == 0:
        return pd.NA

    alpha = (k / (k - 1)) * (1 - (item_variances.sum() / total_variance))
    return alpha

construct_alpha_results = []

for construct, items in construct_items.items():
    pre_cols = ["pre_" + item for item in items]
    post_cols = ["post_" + item for item in items]

    alpha_pre = cronbach_alpha(merged_data[pre_cols])
    alpha_post = cronbach_alpha(merged_data[post_cols])

    construct_alpha_results.append({
        "construct": construct,
        "alpha_pre": round(alpha_pre, 3) if pd.notna(alpha_pre) else pd.NA,
        "alpha_post": round(alpha_post, 3) if pd.notna(alpha_post) else pd.NA
    })

construct_alpha_results = pd.DataFrame(construct_alpha_results)

print("Cronbach alpha results:")
display(construct_alpha_results)
```

```python
# save Cronbach alpha results
construct_alpha_results.to_csv(
    P.tables_dir / "construct_cronbach_alpha_summary.csv",
    index=False
)

print("Saved construct_cronbach_alpha_summary.csv")
print("Shape:", construct_alpha_results.shape)
```

```python
# Shapiro-Wilk test on pre-post construct difference scores

from scipy.stats import shapiro

def format_p_value(p):
    if pd.isna(p):
        return pd.NA
    elif p < 0.001:
        return "<.001"
    else:
        return f"{p:.3f}"

construct_shapiro_results = []

for construct in construct_items.keys():
    pre_col = "pre_" + construct
    post_col = "post_" + construct

    diff_scores = (merged_data[post_col] - merged_data[pre_col]).dropna()

    shapiro_w, shapiro_p_value = shapiro(diff_scores)

    construct_shapiro_results.append({
        "construct": construct,
        "shapiro_w": round(shapiro_w, 3),
        "shapiro_p_value": shapiro_p_value,
        "shapiro_p": format_p_value(shapiro_p_value)
    })

construct_shapiro_results = pd.DataFrame(construct_shapiro_results)

print("Shapiro-Wilk results:")
display(construct_shapiro_results[["construct", "shapiro_w", "shapiro_p"]])
```

```python
# save Shapiro-Wilk results
construct_shapiro_results.to_csv(
    P.tables_dir / "construct_shapiro_normality_summary.csv",
    index=False
)

print("Saved construct_shapiro_normality_summary.csv")
print("Shape:", construct_shapiro_results.shape)
```

```python
# combine construct results with alpha and Shapiro results

construct_results_reliability_normality = (
    construct_results
    .merge(construct_alpha_results, on="construct", how="left")
    .merge(construct_shapiro_results, on="construct", how="left")
)

print("Construct results with reliability and normality:")
display(
    construct_results_reliability_normality[
        [
            "construct",
            "n_complete_pairs",
            "pre_mean",
            "pre_sd",
            "pre_mean_sd",
            "post_mean",
            "post_sd",
            "post_mean_sd",
            "alpha_pre",
            "alpha_post",
            "shapiro_w",
            "shapiro_p"
        ]
    ]
)
```

```python
# save combined reliability and normality summary
construct_results_reliability_normality.to_csv(
    P.tables_dir / "construct_reliability_and_normality_summary.csv",
    index=False
)

print("Saved construct_reliability_and_normality_summary.csv")
print("Shape:", construct_results_reliability_normality.shape)
```

```python
# decide which test to use for each construct
# keep numeric p-values for logic, but show formatted p-values in the table

construct_test_plan = construct_shapiro_results[["construct", "shapiro_p_value", "shapiro_p"]].copy()

construct_test_plan["recommended_test"] = construct_test_plan["shapiro_p_value"].apply(
    lambda p: "paired_t_test" if p > 0.05 else "wilcoxon_signed_rank"
)

print("Recommended test for each construct:")
display(
    construct_test_plan[
        ["construct", "shapiro_p", "recommended_test"]
    ]
)
```

```python
# run paired t-test or Wilcoxon signed-rank based on the Shapiro result

from scipy.stats import ttest_rel, wilcoxon

def format_p_value(p):
    if pd.isna(p):
        return pd.NA
    elif p < 0.001:
        return "<.001"
    else:
        return f"{p:.3f}"

construct_test_results = []

for _, row in construct_test_plan.iterrows():
    construct = row["construct"]
    recommended_test = row["recommended_test"]

    pre_col = "pre_" + construct
    post_col = "post_" + construct

    paired_data = merged_data[[pre_col, post_col]].dropna().copy()
    pre_scores = paired_data[pre_col]
    post_scores = paired_data[post_col]

    if recommended_test == "paired_t_test":
        test_statistic, p_value = ttest_rel(post_scores, pre_scores)
        test_used = "Paired t-test"
        df = len(paired_data) - 1
    else:
        test_statistic, p_value = wilcoxon(post_scores, pre_scores)
        test_used = "Wilcoxon signed-rank"
        df = pd.NA

    construct_test_results.append({
        "construct": construct,
        "test_used": test_used,
        "test_statistic": round(test_statistic, 3),
        "df": df,
        "p_value": format_p_value(p_value)
    })

construct_test_results = pd.DataFrame(construct_test_results)

print("Construct-level test results:")
display(construct_test_results)
```

```python
# save construct-level test results
construct_test_results.to_csv(
    P.tables_dir / "construct_ttest_wilcoxon_summary.csv",
    index=False
)

print("Saved construct_ttest_wilcoxon_summary.csv")
print("Shape:", construct_test_results.shape)
```

```python
# combine means, reliability, normality, and test results

construct_results_with_tests = (
    construct_results_reliability_normality
    .merge(construct_test_results, on="construct", how="left")
)

print("Construct results with tests:")
display(
    construct_results_with_tests[
        [
            "construct",
            "n_complete_pairs",
            "pre_mean",
            "pre_sd",
            "pre_mean_sd",
            "post_mean",
            "post_sd",
            "post_mean_sd",
            "alpha_pre",
            "alpha_post",
            "shapiro_w",
            "shapiro_p",
            "test_used",
            "test_statistic",
            "df",
            "p_value"
        ]
    ]
)
```

```python
# save combined construct results with tests
construct_results_with_tests.to_csv(
    P.tables_dir / "construct_reliability_normality_test_summary.csv",
    index=False
)

print("Saved construct_reliability_normality_test_summary.csv")
print("Shape:", construct_results_with_tests.shape)
```

```python
# paired Cohen's d and 95% bootstrap confidence intervals

import numpy as np

def paired_cohens_d(pre_scores, post_scores):
    diff_scores = post_scores - pre_scores
    diff_sd = diff_scores.std(ddof=1)

    if diff_sd == 0:
        return pd.NA

    return diff_scores.mean() / diff_sd

def interpret_cohens_d(d):
    if pd.isna(d):
        return pd.NA
    elif d < 0.2:
        return "negligible"
    elif d < 0.5:
        return "small"
    elif d < 0.8:
        return "medium"
    else:
        return "large"

construct_effect_size_results = []

rng = np.random.default_rng(42)
n_boot = 5000

for construct in construct_items.keys():
    pre_col = "pre_" + construct
    post_col = "post_" + construct

    paired_data = merged_data[[pre_col, post_col]].dropna().copy()
    pre_scores = paired_data[pre_col].to_numpy()
    post_scores = paired_data[post_col].to_numpy()

    d_value = paired_cohens_d(pre_scores, post_scores)

    boot_d_values = []

    for _ in range(n_boot):
        sample_idx = rng.choice(len(paired_data), size=len(paired_data), replace=True)
        boot_pre = pre_scores[sample_idx]
        boot_post = post_scores[sample_idx]
        boot_d = paired_cohens_d(boot_pre, boot_post)

        if pd.notna(boot_d):
            boot_d_values.append(boot_d)

    d_ci_low, d_ci_high = np.percentile(boot_d_values, [2.5, 97.5])

    construct_effect_size_results.append({
        "construct": construct,
        "cohens_d": round(d_value, 3) if pd.notna(d_value) else pd.NA,
        "d_ci_low": round(d_ci_low, 3),
        "d_ci_high": round(d_ci_high, 3),
        "effect_size_interpretation": interpret_cohens_d(d_value)
    })

construct_effect_size_results = pd.DataFrame(construct_effect_size_results)

print("Construct effect size results:")
display(construct_effect_size_results)
```

```python
# save effect size results

construct_effect_size_results.to_csv(
    P.tables_dir / "construct_paired_cohens_d_ci_summary.csv",
    index=False
)

print("Saved construct_paired_cohens_d_ci_summary.csv")
print("Shape:", construct_effect_size_results.shape)
```

```python
# combine test results with effect sizes

construct_results_final = (
    construct_results_with_tests
    .merge(construct_effect_size_results, on="construct", how="left")
)

print("Final construct results:")
display(
    construct_results_final[
        [
            "construct",
            "n_complete_pairs",
            "pre_mean",
            "pre_sd",
            "pre_mean_sd",
            "post_mean",
            "post_sd",
            "post_mean_sd",
            "alpha_pre",
            "alpha_post",
            "shapiro_w",
            "shapiro_p",
            "test_used",
            "test_statistic",
            "df",
            "p_value",
            "cohens_d",
            "d_ci_low",
            "d_ci_high",
            "effect_size_interpretation"
        ]
    ]
)
```
