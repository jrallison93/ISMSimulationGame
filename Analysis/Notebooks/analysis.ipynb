**Cronbach alpha for each construct at pre and post**

```python
# Cronbach alpha for each construct at pre and post

def cronbach_alpha(df):
    item_scores = df.dropna().copy()
    k = item_scores.shape[1]

    if k < 2:
        return pd.NA

    item_variances = item_scores.var(axis=0, ddof=1)
    total_score = item_scores.sum(axis=1)
    total_variance = total_score.var(ddof=1)

    if total_variance == 0:
        return pd.NA

    alpha = (k / (k - 1)) * (1 - (item_variances.sum() / total_variance))
    return alpha

construct_alpha_results = []

for construct, items in construct_items.items():
    pre_cols = ["pre_" + item for item in items]
    post_cols = ["post_" + item for item in items]

    alpha_pre = cronbach_alpha(merged_data[pre_cols])
    alpha_post = cronbach_alpha(merged_data[post_cols])

    construct_alpha_results.append({
        "construct": construct,
        "alpha_pre": round(alpha_pre, 3) if pd.notna(alpha_pre) else pd.NA,
        "alpha_post": round(alpha_post, 3) if pd.notna(alpha_post) else pd.NA
    })

construct_alpha_results = pd.DataFrame(construct_alpha_results)

print("Cronbach alpha results:")
display(construct_alpha_results)
```

Cronbach alpha values were high across all five constructs at both pre-test and post-test, indicating strong internal consistency. This suggests that the items within each construct worked well together as a coherent scale. The very high alpha values for CIA triad and access control should be interpreted with some caution, as they may reflect closely related items rather than broader construct coverage.

**Shapiro-Wilk test on pre-post construct difference scores**

```python
# Shapiro-Wilk test on pre-post construct difference scores

from scipy.stats import shapiro

def format_p_value(p):
    if pd.isna(p):
        return pd.NA
    elif p < 0.001:
        return "<.001"
    else:
        return f"{p:.3f}"

construct_shapiro_results = []

for construct in construct_items.keys():
    pre_col = "pre_" + construct
    post_col = "post_" + construct

    diff_scores = (merged_data[post_col] - merged_data[pre_col]).dropna()

    shapiro_w, shapiro_p_value = shapiro(diff_scores)

    construct_shapiro_results.append({
        "construct": construct,
        "shapiro_w": round(shapiro_w, 3),
        "shapiro_p_value": shapiro_p_value,
        "shapiro_p": format_p_value(shapiro_p_value)
    })

construct_shapiro_results = pd.DataFrame(construct_shapiro_results)

print("Shapiro-Wilk results:")
display(construct_shapiro_results[["construct", "shapiro_w", "shapiro_p"]])
```

The Shapiro-Wilk results show that the distribution of pre-post difference scores was not normal for most constructs. Only behavioural decision-making confidence met the normality assumption, while cognitive decision-making confidence, CIA triad, access control, and malware and hacking showed non-normal distributions. This indicates that a mixed test-selection approach would lead to different statistical frameworks across constructs, which is why both paired t-tests and Wilcoxon signed-rank tests were later examined as a robustness check.

```python
# save Shapiro-Wilk results
construct_shapiro_results.to_csv(
    P.tables_dir / "construct_shapiro_normality_summary.csv",
    index=False
)

print("Saved construct_shapiro_normality_summary.csv")
print("Shape:", construct_shapiro_results.shape)
```

**decide which test to use for each construct**

```python
# decide which test to use for each construct
# keep numeric p-values for logic, but show formatted p-values in the table

construct_test_plan = construct_shapiro_results[["construct", "shapiro_p_value", "shapiro_p"]].copy()

construct_test_plan["recommended_test"] = construct_test_plan["shapiro_p_value"].apply(
    lambda p: "paired_t_test" if p > 0.05 else "wilcoxon_signed_rank"
)

print("Recommended test for each construct:")
display(
    construct_test_plan[
        ["construct", "shapiro_p", "recommended_test"]
    ]
)
```

Based on the Shapiro-Wilk results, behavioural decision-making confidence met the normality assumption and would therefore be analysed using a paired t-test, while the other four constructs would be analysed using the Wilcoxon signed-rank test. However, because this mixed framework could appear inconsistent across closely related constructs, both tests were later run across all five constructs as a robustness check.

**Paired t-test or Wilcoxon signed-rank based on the Shapiro result**

```python
# run paired t-test or Wilcoxon signed-rank based on the Shapiro result

from scipy.stats import ttest_rel, wilcoxon

def format_p_value(p):
    if pd.isna(p):
        return pd.NA
    elif p < 0.001:
        return "<.001"
    else:
        return f"{p:.3f}"

construct_test_results = []

for _, row in construct_test_plan.iterrows():
    construct = row["construct"]
    recommended_test = row["recommended_test"]

    pre_col = "pre_" + construct
    post_col = "post_" + construct

    paired_data = merged_data[[pre_col, post_col]].dropna().copy()
    pre_scores = paired_data[pre_col]
    post_scores = paired_data[post_col]

    if recommended_test == "paired_t_test":
        test_statistic, p_value = ttest_rel(post_scores, pre_scores)
        test_used = "Paired t-test"
        df = len(paired_data) - 1
    else:
        test_statistic, p_value = wilcoxon(post_scores, pre_scores)
        test_used = "Wilcoxon signed-rank"
        df = pd.NA

    construct_test_results.append({
        "construct": construct,
        "test_used": test_used,
        "test_statistic": round(test_statistic, 3),
        "df": df,
        "p_value": format_p_value(p_value)
    })

construct_test_results = pd.DataFrame(construct_test_results)

print("Construct-level test results:")
display(construct_test_results)
```

Using the assumption-based test selection, all five constructs showed statistically significant pre-post improvement. Behavioural decision-making confidence was tested with a paired t-test, while the remaining four constructs were tested with Wilcoxon signed-rank. In every case, the results indicate that post-test scores were significantly higher than pre-test scores.

**Paired Cohen's d and 95% bootstrap confidence intervals**

```python
# paired Cohen's d and 95% bootstrap confidence intervals

import numpy as np

def paired_cohens_d(pre_scores, post_scores):
    diff_scores = post_scores - pre_scores
    diff_sd = diff_scores.std(ddof=1)

    if diff_sd == 0:
        return pd.NA

    return diff_scores.mean() / diff_sd

def interpret_cohens_d(d):
    if pd.isna(d):
        return pd.NA
    elif d < 0.2:
        return "negligible"
    elif d < 0.5:
        return "small"
    elif d < 0.8:
        return "medium"
    else:
        return "large"

construct_effect_size_results = []

rng = np.random.default_rng(42)
n_boot = 5000

for construct in construct_items.keys():
    pre_col = "pre_" + construct
    post_col = "post_" + construct

    paired_data = merged_data[[pre_col, post_col]].dropna().copy()
    pre_scores = paired_data[pre_col].to_numpy()
    post_scores = paired_data[post_col].to_numpy()

    d_value = paired_cohens_d(pre_scores, post_scores)

    boot_d_values = []

    for _ in range(n_boot):
        sample_idx = rng.choice(len(paired_data), size=len(paired_data), replace=True)
        boot_pre = pre_scores[sample_idx]
        boot_post = post_scores[sample_idx]
        boot_d = paired_cohens_d(boot_pre, boot_post)

        if pd.notna(boot_d):
            boot_d_values.append(boot_d)

    d_ci_low, d_ci_high = np.percentile(boot_d_values, [2.5, 97.5])

    construct_effect_size_results.append({
        "construct": construct,
        "cohens_d": round(d_value, 3) if pd.notna(d_value) else pd.NA,
        "d_ci_low": round(d_ci_low, 3),
        "d_ci_high": round(d_ci_high, 3),
        "effect_size_interpretation": interpret_cohens_d(d_value)
    })

construct_effect_size_results = pd.DataFrame(construct_effect_size_results)

print("Construct effect size results:")
display(construct_effect_size_results)
```

The effect size results suggest that the strongest improvements were in cognitive and behavioural decision-making confidence, both of which showed large effects. Malware and hacking and access control showed medium effects, while the CIA triad construct showed a smaller but still meaningful effect. Overall, these results indicate that the intervention had its greatest impact on decision-making confidence, with more moderate gains in knowledge-based areas.

## Robustness Check: Paired t-test and Wilcoxon Signed-Rank Test for All Constructs

**Run both tests for all 5 constructs and save the table**

```python
#run both tests for all 5 constructs and save the table
from scipy.stats import ttest_rel, wilcoxon
import pandas as pd
import numpy as np

def format_p_value_dual(p):
    if pd.isna(p):
        return pd.NA
    elif p < 0.001:
        return "<.001"
    else:
        return f"{p:.3f}"

construct_dual_test_results = []

for construct in construct_items.keys():
    pre_col = "pre_" + construct
    post_col = "post_" + construct

    paired_data = merged_data[[pre_col, post_col]].dropna().copy()
    pre_scores = paired_data[pre_col]
    post_scores = paired_data[post_col]

    paired_t_stat, paired_t_p = ttest_rel(post_scores, pre_scores)
    paired_t_df = len(paired_data) - 1

    wilcoxon_stat, wilcoxon_p = wilcoxon(post_scores, pre_scores)

    construct_dual_test_results.append({
        "construct": construct,
        "n_complete_pairs": len(paired_data),
        "paired_t_statistic": round(paired_t_stat, 3),
        "paired_t_df": paired_t_df,
        "paired_t_p_value": format_p_value_dual(paired_t_p),
        "wilcoxon_w_statistic": round(wilcoxon_stat, 3),
        "wilcoxon_p_value": format_p_value_dual(wilcoxon_p)
    })

construct_dual_test_results = pd.DataFrame(construct_dual_test_results)

print("Paired t-test and Wilcoxon results:")
display(construct_dual_test_results)

construct_dual_test_results.to_csv(
    P.tables_dir / "Robustness_Check_construct_paired_ttest_and_wilcoxon_results.csv",
    index=False
)

print("Saved construct_paired_ttest_and_wilcoxon_results.csv")
print("Shape:", construct_dual_test_results.shape)
```

As a robustness check, both paired t-tests and Wilcoxon signed-rank tests were run across all five constructs using the 72 matched pairs. Both testing approaches produced the same overall pattern, with statistically significant pre-post improvement observed for every construct. This suggests that the findings are stable and do not depend on whether a parametric or non-parametric framework is used.

**Effect sizes for both test frameworks, for all 5 constructs**

```python
# effect sizes for both test frameworks, for all 5 constructs
from scipy.stats import wilcoxon, rankdata
import numpy as np
import pandas as pd

def interpret_d_clean(d):
    if pd.isna(d):
        return pd.NA
    elif d < 0.2:
        return "negligible"
    elif d < 0.5:
        return "small"
    elif d < 0.8:
        return "medium"
    else:
        return "large"

def interpret_r_clean(r):
    if pd.isna(r):
        return pd.NA
    elif r < 0.1:
        return "negligible"
    elif r < 0.3:
        return "small"
    elif r < 0.5:
        return "medium"
    else:
        return "large"

def paired_cohens_d_clean(pre_scores, post_scores):
    diff_scores = post_scores - pre_scores
    diff_sd = diff_scores.std(ddof=1)

    if diff_sd == 0:
        return pd.NA

    return diff_scores.mean() / diff_sd

def wilcoxon_z_r_clean(pre_scores, post_scores):
    pre_scores = np.asarray(pre_scores)
    post_scores = np.asarray(post_scores)

    try:
        result = wilcoxon(
            post_scores,
            pre_scores,
            zero_method="wilcox",
            correction=False,
            alternative="two-sided",
            method="approx"
        )
    except TypeError:
        result = wilcoxon(
            post_scores,
            pre_scores,
            zero_method="wilcox",
            correction=False,
            alternative="two-sided",
            mode="approx"
        )

    w_stat = float(result.statistic)
    p_value = float(result.pvalue)
    z_stat = getattr(result, "zstatistic", None)

    diff_scores = post_scores - pre_scores
    diff_scores = diff_scores[diff_scores != 0]
    n_nonzero = len(diff_scores)

    if z_stat is None or pd.isna(z_stat):
        abs_diff = np.abs(diff_scores)
        ranks = rankdata(abs_diff, method="average")
        w_plus = ranks[diff_scores > 0].sum()

        mean_w = n_nonzero * (n_nonzero + 1) / 4

        _, tie_counts = np.unique(abs_diff, return_counts=True)
        var_w = (
            n_nonzero * (n_nonzero + 1) * (2 * n_nonzero + 1)
            - np.sum(tie_counts * (tie_counts + 1) * (2 * tie_counts + 1))
        ) / 24

        z_stat = (w_plus - mean_w) / np.sqrt(var_w)

    r_value = abs(z_stat) / np.sqrt(n_nonzero) if n_nonzero > 0 else pd.NA

    return w_stat, p_value, float(z_stat), float(r_value)

construct_dual_effect_results = []

rng = np.random.default_rng(42)
n_boot = 5000

for construct in construct_items.keys():
    pre_col = "pre_" + construct
    post_col = "post_" + construct

    paired_data = merged_data[[pre_col, post_col]].dropna().copy()
    pre_scores = paired_data[pre_col].to_numpy()
    post_scores = paired_data[post_col].to_numpy()

    # paired t-test effect size
    d_value = paired_cohens_d_clean(pre_scores, post_scores)

    boot_d_values = []

    for _ in range(n_boot):
        sample_idx = rng.choice(len(paired_data), size=len(paired_data), replace=True)
        boot_pre = pre_scores[sample_idx]
        boot_post = post_scores[sample_idx]
        boot_d = paired_cohens_d_clean(boot_pre, boot_post)

        if pd.notna(boot_d):
            boot_d_values.append(boot_d)

    d_ci_low, d_ci_high = np.percentile(boot_d_values, [2.5, 97.5])

    # wilcoxon effect size
    _, _, z_stat, r_value = wilcoxon_z_r_clean(pre_scores, post_scores)

    construct_dual_effect_results.append({
        "construct": construct,
        "paired_t_cohens_d": round(d_value, 3),
        "paired_t_d_ci_low": round(d_ci_low, 3),
        "paired_t_d_ci_high": round(d_ci_high, 3),
        "paired_t_d_interpretation": interpret_d_clean(d_value),
        "wilcoxon_z_statistic": round(z_stat, 3),
        "wilcoxon_r": round(r_value, 3),
        "wilcoxon_r_interpretation": interpret_r_clean(r_value)
    })

construct_dual_effect_results = pd.DataFrame(construct_dual_effect_results)

print("Effect size results for both test frameworks:")
display(construct_dual_effect_results)

construct_dual_effect_results.to_csv(
    P.tables_dir / "Robustness_Check_construct_effect_sizes_ttest_and_wilcoxon.csv",
    index=False
)

print("Saved Robustness_Check_construct_effect_sizes_ttest_and_wilcoxon.csv")
print("Shape:", construct_dual_effect_results.shape)
```

The effect size comparison across both statistical frameworks shows a consistent pattern of meaningful improvement across all five constructs. Under the paired t-test framework, the largest effects were again found for cognitive and behavioural decision-making confidence, while CIA triad, access control, and malware and hacking showed smaller to moderate effects. Under the Wilcoxon framework, all constructs showed large rank-based effect sizes, indicating that the positive pre-post change remains strong even when evaluated using a non-parametric approach.
