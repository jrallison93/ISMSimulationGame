```python
import sys
from pathlib import Path

def find_project_root(markers=("src", "Data", "Notebooks")):
    current_path = Path.cwd().resolve()

    for path in [current_path] + list(current_path.parents):
        has_all_markers = all((path / marker).exists() for marker in markers)
        if has_all_markers:
            return path

    raise RuntimeError(
        f"Project root could not be found from: {current_path}"
    )

project_root = find_project_root()
src_path = project_root / "src"

if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))

import paths as P
```

```python
# create prefixed versions for merging
# keep student number, email, and consent_to_proceed only once
# drop email and consent_to_proceed from post_clean before merging

student_col = "please_provide_your_student_number_please_note_this_is_only_used_to_enable_matching_of_the_pre_and_postintervention_survey_data_enable_matching_of_the_pre_and_postintervention_survey_data"

keep_once_cols = [student_col, "email", "consent_to_proceed"]

pre_merge = pre_clean.rename(
    columns={col: f"pre_{col}" for col in pre_clean.columns if col not in keep_once_cols}
).copy()

post_merge = post_clean.drop(columns=["email", "consent_to_proceed"]).copy()

post_merge = post_merge.rename(
    columns={col: f"post_{col}" for col in post_merge.columns if col != student_col}
)

print("Pre merge columns:")
print(pre_merge.columns.tolist())
print()

print("Post merge columns:")
print(post_merge.columns.tolist())
```

```python
# merge using exact matching student numbers only
# keep student number, email, and consent_to_proceed once
# place pre and post versions side by side

merged_data = pd.merge(
    pre_merge,
    post_merge,
    on=student_col,
    how="inner"
).copy()

ordered_cols = [student_col, "email", "consent_to_proceed"]

pre_cols = [
    col for col in pre_merge.columns
    if col not in [student_col, "email", "consent_to_proceed"]
]

post_cols = [
    col for col in post_merge.columns
    if col != student_col
]

pre_base_names = [col.replace("pre_", "", 1) for col in pre_cols]
post_base_names = [col.replace("post_", "", 1) for col in post_cols]

for base_name in pre_base_names:
    pre_col = "pre_" + base_name
    post_col = "post_" + base_name

    if pre_col in merged_data.columns:
        ordered_cols.append(pre_col)
    if post_col in merged_data.columns:
        ordered_cols.append(post_col)

for base_name in post_base_names:
    post_col = "post_" + base_name
    if post_col not in ordered_cols:
        ordered_cols.append(post_col)

merged_data = merged_data[ordered_cols]

print("Merged dataset shape:")
print(merged_data.shape)
print()

print("First 20 merged columns:")
print(merged_data.columns.tolist()[:20])
```

```python
# check matched and unmatched student numbers after merge
match_check = pd.merge(
    pre_clean[[student_col]],
    post_clean[[student_col]],
    on=student_col,
    how="outer",
    indicator=True
)

print("Match summary:")
print(match_check["_merge"].value_counts())
```
